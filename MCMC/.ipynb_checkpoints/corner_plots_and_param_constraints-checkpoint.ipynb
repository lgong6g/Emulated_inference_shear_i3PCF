```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
from scipy import interpolate
from chainconsumer import ChainConsumer
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
data_xi = np.load('MCMC_2PCF_cosmogrid_14dim_model_1e3steps_alphaIA_0_w_nside512_updated_H_chi_D.npy')[:,:5]
data_xi_iZ = np.load('MCMC_all_data_vector_cosmogrid_14dim_model_1e3steps_alphaIA_0_w_nside512_updated_H_chi_D.npy')[:,:5]
#data_xi_iZ_noIA = np.load('MCMC_all_data_vector_cosmogrid_13dim_model_1e3steps_w_nside512_systematics_nz_m_updated_H_chi.npy')[:,:5]
```

```python
c = ChainConsumer()
c.add_chain(data_xi, parameters=["$\Omega_m$", r"${\rm ln}(10^{10} A_s)$", "$w_0$", "$h$", r"$c_{\rm min}$"], name=r'$\xi_{\pm}$', shade=True, shade_alpha=0.7, color='g', linestyle='-', linewidth=2.0, kde=False)
c.add_chain(data_xi_iZ, parameters=["$\Omega_m$", r"${\rm ln}(10^{10} A_s)$", "$w_0$", "$h$", r"$c_{\rm min}$"], name=r'$\xi_{\pm} + \zeta_{\pm, \{50^{\prime}\}-\{130^{\prime}\}}$', shade=True, shade_alpha=0.1, color='r', linestyle='--', linewidth=2.0, kde=False)
#c.add_chain(data_xi_iZ_noIA, parameters=["$\Omega_m$", r"${\rm ln}(10^{10} A_s)$", "$w_0$", "$h$", r"$c_{\rm min}$"], name=r'$\xi_{\pm} + \zeta_{\pm, \{50^{\prime}\}-\{130^{\prime}\}}$'+' no IA', shade=True, shade_alpha=0.5, color='g', linestyle='-', linewidth=2.0, kde=False)
c.configure_truth(linestyle='dotted', linewidth=0.5)
c.configure(label_font_size=20.0, tick_font_size=16.0, max_ticks=4, summary=False, spacing=2.0, legend_kwargs={'fontsize':30})
```

```python
fig = c.plotter.plot(figsize=(10, 10), truth=[0.26, np.log(10**10*2.95098e-09), -1.0, 0.673, 3.13], legend=True)
```
